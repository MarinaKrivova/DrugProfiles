```python
df_328 = df[df["DRUG_ID"]==328].copy()
```

```python
cell_features = pd.read_csv(_FOLDER+"Cell_Line_Features_PANCAN_simple_MOBEM.tsv", sep="\t")
cell_features.rename(columns={"Unnamed: 0": "genes"}, inplace = True)
```

```python
def swap_rows_columns(df, indexes_first_column = True):
    if indexes_first_column:
        df_transfromed = pd.DataFrame(data= df[df.columns[1:]].values.T,
                          index= df.columns[1:], columns= df[df.columns[0]].values)
    else:
        print("The logic of this function is not applicable")
        
    return df_transfromed 
```

```python
cell_features_T = swap_rows_columns(cell_features)

cell_features_T.index = np.array(cell_features_T.index, dtype = "int")
print("Number of drugs in filtered dataset:", df_328["COSMIC_ID"].nunique())

# Not all the drugs from filtered dataset are present in cell lines features
common_cells_drug328 = list(set(np.array(cell_features.columns[1:], dtype="int"))& set(df_328["COSMIC_ID"].values))
print("Number of common drugs in both datasets", len(common_cells_drug328))

cell_328 = cell_features_T.loc[common_cells_drug328, :].reset_index()
cell_328.rename(columns = {"index": "COSMIC_ID"}, inplace=True)
```

```python
param_cols = []
df= df_328.copy()
for i in range(4):
    param_col = "param_"+str(i+1)
    param_cols.append(param_col)
    df[param_col] = df['fitting_param_s4'].apply(lambda x: x[i])

param_cols.insert(0, "COSMIC_ID")
```

```python
# merge filtered data and cell line features

merged_df = pd.merge(left=df[param_cols], right = cell_328, on = "COSMIC_ID")
```

```python
merged_df[merged_df.columns[:4]]
```

```python
%%time
X = merged_df[merged_df.columns[5:]].values
scores = {}
for i in range(4):
    y = merged_df["param_"+str(i+1)]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)
    kr_lin = KernelRidge(kernel='linear')
    kr_lin.fit(X_train, y_train)
    y_pred = kr_lin.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    scores["param_"+str(i+1)] = mse
    print(mse)
```
